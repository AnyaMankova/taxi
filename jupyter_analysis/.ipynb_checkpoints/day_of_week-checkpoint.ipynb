```python
import pandas as pd
import matplotlib.pyplot as plt
uber_df = pd.read_csv("uber-raw-data-sep14.csv")
#преобразовываю данные в стандартный формат (изначально было - 9/1/2014)
uber_df['Date/Time'] = pd.to_datetime(uber_df['Date/Time'])
uber_df['Day_of_week'] = uber_df['Date/Time'].dt.day_name()

#print(uber_df.head(5))

#создаю фигуру и добавляю к ней область рисования
fig = plt.figure(figsize=[12, 6], facecolor='white')
ax = fig.add_subplot()

ax.hist(uber_df['Day_of_week'], width= 0.6, range= (0, 6.5), bins=7, color="orange", edgecolor='black')
ax.set_title('Распределение вызовов такси по дням недели', fontsize=16)
ax.set_xlabel('День недели', fontsize=14)
#ax.set_xticks(['Понедельник', 'Вторник', 'Среда', 'Четверг', 'Пятница', 'Суббота', 'Воскресенье'])
ax.set_ylabel('Количество вызовов', fontsize=14)
plt.savefig('images/Day_of_week.png', dpi=120)
plt.show()
```
