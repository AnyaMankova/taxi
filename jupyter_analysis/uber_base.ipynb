```python
import pandas as pd
import matplotlib.pyplot as plt
uber_df = pd.read_csv('uber-raw-data-sep14.csv')

# подсчитываем количество станций
count_base = uber_df.Base.nunique()
#создаю фигуру и добавляю к ней область рисования
fig = plt.figure(figsize=[12, 6], facecolor='white')
ax = fig.add_subplot()

ax.hist(uber_df['Base'], bins=count_base, color= "orange", edgecolor='black')
ax.set_title('Распределение вызовов такси по станциям Uber', fontsize=16)
ax.set_xlabel('Станция', fontsize=14)
ax.set_ylabel('Количество вызовов', fontsize=14)
plt.savefig('images/Uber_base.png', dpi=120)
plt.show()
```
